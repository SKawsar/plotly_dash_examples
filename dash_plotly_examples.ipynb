```python
import pandas as pd
import plotly.express as px
```

```python
ecom_sales = pd.read_csv("ecom_sales.csv")

print(ecom_sales.shape)
display(ecom_sales.sample(5))
```

```python
ecom_sales = ecom_sales.groupby('Country')['OrderValue'].agg('sum')
ecom_sales = ecom_sales.reset_index(name='Total Sales ($)')

print(ecom_sales.shape)
display(ecom_sales.head())
```

```python
# Create the bar chart
# Set the DataFrame, x and y
# Set the graph to be horizontal
bar_fig = px.bar(data_frame=ecom_sales, 
                 x='Total Sales ($)', 
                 y='Country',
                 orientation='h', 
                 title='Total Sales by Country')

# Increase the gap between bars
bar_fig.update_layout({'bargap': 0.5})

bar_fig.show()
```
